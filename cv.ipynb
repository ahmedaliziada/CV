```python
captures = cv2.VideoCapture('dog.mp4')
```

```python
isTrue, cap = captures.read()
```

```python
while isTrue:
    isTrue, frames = captures.read()

    if not isTrue:
        break
    
    frames_resized = cv2.resize(frames, (640, 480))
    cv2.imshow('Video',frames_resized )
    
    
    if cv2.waitKey(20) & 0xFF==ord('q'):
        break
    
    
captures.release()
cv2.destroyAllWindows()
```

```python
while isTrue:
    isTrue, frames = captures.read()

    if not isTrue:
        captures.set(cv2.CAP_PROP_POS_FRAMES, 0)
        break
    
    frames_resized = cv2.resize(frames, (640, 480))
    cv2.imshow('Video', frames_resized)
    

    cv2.waitKey(20)
    
    if cv2.waitKey(20) & 0xFF == ord('q'):
        break

captures.release()
cv2.destroyAllWindows()
```
